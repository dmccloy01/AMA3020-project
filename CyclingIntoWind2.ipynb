```python
import numpy as np

#define constraints
rho = 1.2250
c_d = 1.1
A = 0.5

#define function for air resistance
def air_resistance(v_c, v_w, theta):
  v2 = v_c**2 + v_w**2 - 2*v_c*v_w*np.cos(np.radians(theta))
  return 0.5*rho*c_d*A*v2

#constant cyclist speed 25kph
v_c = 7

#define wind speeds
v_w_vals = np.linspace(0, 20, 1000)
v_w_vals_kph = v_w_vals*3.6

#calculate air resistance for tailwind
tailwind_speed = air_resistance(v_c, v_w_vals, 0)

#calculate air resistance for headwind
headwind_speed = air_resistance(v_c, v_w_vals, 180)

#calculate air resistance for crosswind
crosswind_speed = air_resistance(v_c, v_w_vals, 90)

#plot results
import matplotlib.pyplot as plt
plt.plot(v_w_vals_kph, tailwind_speed, label='Tailwind')
plt.plot(v_w_vals_kph, headwind_speed, label='Headwind')
plt.plot(v_w_vals_kph, crosswind_speed, label='Crosswind')
plt.xlabel('Wind Speed (kph)')
plt.ylabel('Air Resistance (N)')
plt.title('Air Resistance vs. Wind Speed')
plt.legend()
plt.grid(True)
plt.savefig('air_resistance.pdf')
plt.show()

#define wind angles
theta_vals = np.linspace(0, 360, 1000)

#calculate air resistance for 10m/s wind
wind_speed = air_resistance(v_c, 10, theta_vals)

#plot results
plt.plot(theta_vals, wind_speed)
plt.title('Air Resistance vs. Wind Angle for a 36kph Wind')
plt.xlabel('Wind Angle (degrees)')
plt.ylabel('Air Resistance (N)')
plt.grid(True)
plt.savefig('air_resistance_20m.pdf')
plt.show()

#values at 20ms
print(air_resistance(v_c, 20, 0))
print(air_resistance(v_c, 20, 180))

#define varying cyclist speed upto 45 kph
v_c_vals = np.linspace(0, 12.5, 1000)
v_c_vals_kph = v_c_vals*3.6

#calculate air resistance for varying cyclist speed for 7m/s wind
varying_cyc_speed_hw = air_resistance(v_c_vals, 7, 180)
varying_cyc_speed_tw = air_resistance(v_c_vals, 7, 0)

#plot results
plt.plot(v_c_vals_kph, varying_cyc_speed_hw, label = 'Headwind')
plt.plot(v_c_vals_kph, varying_cyc_speed_tw, label = 'Tailwind')
plt.title('Air Resistance vs. Cyclist Speed for a 25kph Wind')
plt.xlabel('Cyclist Speed (kph)')
plt.ylabel('Air Resistance (N)')
plt.grid(True)
plt.legend()
plt.savefig('air_resistance_varying_cyc_speed.pdf')
plt.show()

#define constants
Crr = 0.0025
mass = 80
g = 9.81

#calculate air resistance
air_res = air_resistance(v_c_vals, 0, 180)
#calculate rolling resistance
rr = Crr * mass * g * np.ones_like(v_c_vals)

#plot rolling resistance and air resistance vs cyclist speed
plt.plot(v_c_vals_kph, rr, label = 'Rolling Resistance')
plt.plot(v_c_vals_kph, air_res, label = 'Air Resistance')
plt.title('Air Resistance vs. Rolling Resistance')
plt.xlabel('Cyclist Speed (kph)')
plt.ylabel('Drag Force (N)')
plt.grid(True)
plt.legend()
plt.savefig('air_resistance_vs_rr.pdf')
plt.show()

```
